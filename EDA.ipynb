What is the probability that a student is a debtor given that he is an international student VS Dual-Citizen?
Total students = 4425

```python
prob_international_std = Original_df['International'].eq(1).sum()
prob_international_std = prob_international_std/4425
prob_dual_citizen = Original_df['Dual-Citizen'].eq(1).sum()
prob_dual_citizen = prob_dual_citizen/4425 
print(f"DUAL:{prob_dual_citizen}")
print(f"INTERNATIONAL:{prob_international_std}")
```

Debtor given international

```python
p_debtor = Original_df['Debtor'].eq(1).sum()
p_debtor = p_debtor/4425
p_debtor_international = (Original_df['Debtor'].eq(1) & Original_df['International'].eq(1)).sum() /4425

final1 = p_debtor_international/p_debtor
print(f"probability that a student is a debtor given that he is a international student is : {final1}")
```

Debtor given dual-citizen

```python
p_debtor_dual = (Original_df['Debtor'].eq(1) & Original_df['Dual-Citizen'].eq(1)).sum() /4425
final2 = p_debtor_dual = p_debtor
print(f"probability that a student is a debtor given that he is a dual-citizen is : {final2}")
```

Means that a international student is more likely to take scholarship or financial aid as loans might not be available to him or he is much less likely to prefer loans but lets further verify

Probability of a international students with scholarships compared to dual citizens

```python
international_scholar = (Original_df['International'].eq(1) & Original_df['Scholarship holder']).sum() / 4425
dual_citizen_scholar = (Original_df['Dual-Citizen'].eq(1) & Original_df['Scholarship holder']).sum() / 4425

print(international_scholar > dual_citizen_scholar)
```

Shows our intitial analysis is false, my explanation to this is that there are many dual-citizen who are international students somewhere so the the 2 classes we were comparing aren't disjoint sets.

```python
data_temp = {'Category': ['International Scholar', 'Dual Citizen Scholar'], 'Proportion': [international_scholar, dual_citizen_scholar]}
plt.bar(data_temp['Category'],data_temp['Proportion'])
# Add axis labels and title
plt.xlabel('Classes')
plt.ylabel('proportion')
plt.title('International VS Dual Citizen')

# Show the chart
plt.show()
```
